### 2.1 技术原理对比

| Dimension | scRNA-seq | Spatial Transcriptomics |
|-----------|-----------|------------------------|
| **Sample preparation** | Dissociate tissue → single cells | Keep tissue intact |
| **Spatial information** | Lost | Preserved |
| **Resolution** | Single-cell | Spot (10-55μm) or single-cell |
| **Gene coverage** | Whole transcriptome | Whole transcriptome or targeted |
| **Cell number** | 10³-10⁶ cells | 10³-10⁴ spots/cells |
| **Cost per cell** | Low ($0.01-0.1) | Medium to high ($0.1-1) |
| **Tissue context** | No | Yes |

### 2.2 分析流程对比

| Analysis step | scRNA-seq | Spatial Transcriptomics |
|---------------|-----------|------------------------|
| **Quality control** | nUMI, nGene, %MT | + Spatial distribution |
| **Normalization** | Library size | + Spatial smoothing |
| **Dimensionality reduction** | PCA, UMAP | + Spatial PCA |
| **Clustering** | Graph-based | + Spatial constraint |
| **Marker genes** | Differential expression | + Spatially variable genes |
| **Cell type annotation** | Reference-based | + Deconvolution |
| **Cell-cell interaction** | Ligand-receptor database | + Spatial proximity |
| **Trajectory inference** | Pseudotime | + Spatial trajectory |

### 2.3 应用场景对比

| Research question | scRNA-seq | ST | Why ST is better |
|-------------------|-----------|----|--------------------|
| **Cell type discovery** | ✓✓✓ | ✓✓ | scRNA has higher throughput |
| **Rare cell detection** | ✓✓✓ | ✓ | scRNA can profile more cells |
| **Tissue architecture** | ✗ | ✓✓✓ | ST preserves spatial context |
| **Cell-cell communication** | ✓ | ✓✓✓ | ST validates physical proximity |
| **Tumor microenvironment** | ✓ | ✓✓✓ | ST reveals spatial organization |
| **Developmental biology** | ✓✓ | ✓✓✓ | ST captures spatial gradients |
| **Niche identification** | ✗ | ✓✓✓ | ST defines microenvironment |
| **Spatial biomarker** | ✗ | ✓✓✓ | ST links expression to location |

**核心差异**：
- scRNA-seq 适合"发现细胞类型"和"大规模筛选"
- ST 适合"理解组织结构"和"验证空间假设"
- 最佳策略：scRNA-seq 先探索，ST 后验证

**ST 数据**：

ST 数据比 scRNA-seq 多了三个关键组件：
1. **空间坐标**：每个 spot/细胞的位置

## 第三部分：ST 应用图谱

### 3.1 肿瘤学：从"细胞图谱"到"战场地图"

**传统 scRNA-seq 研究**：
- 肿瘤中有哪些免疫细胞？
- 哪些基因在肿瘤细胞中高表达？
- T 细胞有哪些亚型？

**ST 带来的新视角**：
- 免疫细胞在肿瘤的哪个位置？（中心 vs 边缘 vs 基质）
- 肿瘤-免疫界面发生了什么？（免疫浸润 vs 免疫排斥）
- 哪些区域对免疫治疗敏感？（空间生物标志物）

**典型应用**：
1. **肿瘤微环境分区**
   - 识别肿瘤中心、侵袭前沿、免疫浸润区、坏死区
   - 每个区域的细胞组成和基因表达特征
   - 预测治疗响应的空间标志物

2. **免疫逃逸机制**
   - 免疫细胞是否接触肿瘤细胞？
   - 免疫抑制信号在哪里表达？
   - 物理屏障（如基质）如何阻止免疫浸润？

3. **转移前生态位**
   - 肿瘤细胞如何准备转移？
   - 转移前生态位的空间特征
   - 血管和淋巴管周围的细胞状态

### 3.2 神经科学：从"细胞类型"到"脑区地图"

**传统 scRNA-seq 研究**：
- 大脑有多少种神经元类型？
- 不同脑区的基因表达差异
- 神经退行性疾病中哪些细胞受影响？

**ST 带来的新视角**：
- 神经元如何在空间上组织？（层状结构、核团）
- 神经回路的空间连接模式
- 疾病如何在空间上传播？（如阿尔茨海默病的 tau 蛋白扩散）

**典型应用**：
1. **脑区功能分区**
   - 皮层六层结构的分子特征
   - 海马、杏仁核等核团的边界
   - 功能区的空间标志物

2. **神经退行性疾病**
   - 病理蛋白（如 Aβ、tau）的空间分布
   - 神经元丢失的空间模式
   - 炎症和胶质细胞激活的空间关系

3. **发育神经生物学**
   - 神经干细胞的空间分布
   - 神经元迁移路径
   - 脑区发育的时空动态

### 3.3 发育生物学：从"时间轴"到"时空地图"

**传统 scRNA-seq 研究**：
- 细胞分化的轨迹（pseudotime）
- 不同发育阶段的基因表达变化
- 关键转录因子的作用

**ST 带来的新视角**：
- 细胞分化在空间上如何发生？（梯度 vs 边界）
- 组织形态发生的空间机制
- 信号中心（如 organizer）如何影响周围细胞？

**典型应用**：
1. **器官发育**
   - 肝脏、肾脏、心脏的空间分区形成
   - 形态发生素梯度的空间分布
   - 细胞迁移和重排的空间模式

2. **胚胎发育**
   - 三胚层分化的空间组织
   - 体节形成的时空动态
   - 神经管闭合的空间机制

3. **再生医学**
   - 组织再生的空间模式
   - 干细胞生态位的空间特征
   - 再生信号的空间传播

### 3.4 免疫学：从"细胞状态"到"免疫地形"

**传统 scRNA-seq 研究**：
- 免疫细胞的激活状态
- T 细胞、B 细胞、巨噬细胞的亚型
- 细胞因子和趋化因子的表达

- 免疫反应在哪里发生？（淋巴滤泡 vs 弥漫浸润）
- 免疫细胞如何迁移和聚集？

**典型应用**：
1. **淋巴器官结构**
   - 淋巴结的 T 区、B 区、生发中心
   - 脾脏的白髓、红髓分区
   - 胸腺的皮质、髓质结构

2. **炎症和感染**
   - 炎症灶的空间组织
   - 病原体感染的空间模式
   - 免疫细胞募集的空间动态

3. **自身免疫疾病**
   - 自身抗体产生的空间位置
   - 组织损伤的空间模式
   - 免疫调节失衡的空间特征

## 第四部分：为什么是"下一代主战场"

### 4.1 技术成熟度曲线

空间转录组正处于技术成熟度曲线的"爬坡期"：

**2016-2019：概念验证期**
- Visium 等平台推出
- 分辨率低（55μm spot，包含 5-20 个细胞）
- 工具不成熟，分析困难

**2020-2022：快速发展期**
- 高分辨率平台出现（Visium HD、Xenium、CosMx）
- 分析工具爆发（BayesSpace、Cell2location、Squidpy）
- 应用场景扩展（肿瘤、神经、发育）

**2023-2025：成熟应用期**
- 单细胞分辨率成为标准
- 工具生态完善（Scanpy + Squidpy 成为主流）
- 临床转化开始（空间生物标志物）

**2026+：主流技术期**
- 成本降低，通量提高
- 多组学整合（ST + 蛋白 + 代谢）
- 成为标准实验手段（如 scRNA-seq 在 2018 年后的地位）

### 4.2 市场和投资趋势

**商业平台竞争激烈**：
- 10x Genomics（Visium、Visium HD）
- NanoString（CosMx）
- 10x Genomics（Xenium，收购自 Akoya）
- Vizgen（MERFISH）
- BGI（Stereo-seq）

**投资持续增长**：
- 2020-2025 年，空间组学公司融资超过 10 亿美元
- 大型制药公司建立 ST 平台（如辉瑞、罗氏）
- 临床诊断公司布局空间病理

**学术产出爆发**：
- 2020 年：~200 篇 ST 相关论文
- 2023 年：~1000 篇
- 2025 年：~2000 篇（预计）
- Nature、Science、Cell 高频发表 ST 研究

### 4.3 未解决的挑战

ST 成为主战场，但仍有挑战：

**技术挑战**：
1. **分辨率 vs 通量权衡**
   - 高分辨率（单细胞）→ 视野小、成本高
   - 低分辨率（spot）→ 需要反卷积，结果不确定

2. **基因覆盖 vs 分辨率权衡**
   - 全转录组（Visium）→ 分辨率低
   - 靶向基因（Xenium）→ 基因数有限

3. **样本类型限制**
   - 新鲜冷冻 vs FFPE
   - 组织厚度、大小限制

**分析挑战**：
1. **反卷积不确定性**
   - spot 级数据需要推断细胞组成
   - 不同工具结果差异大
   - 缺乏金标准验证

2. **空间统计方法**
   - 如何定义"空间显著性"？
   - 如何校正多重检验？
   - 如何处理空间自相关？

3. **多样本整合**
   - 批次效应更复杂（技术 + 空间）
   - 如何对齐不同样本的空间坐标？
   - 如何比较不同组织的空间模式？

**生物学挑战**：
1. **因果推断**
   - 空间共定位不等于因果关系
   - 如何区分直接作用 vs 间接作用？
   - 如何验证细胞通讯的功能？

2. **动态过程**
   - ST 是静态快照，如何推断动态过程？
   - 如何结合时间序列和空间信息？

3. **3D 重建**
   - 2D 切片如何重建 3D 结构？
   - 如何对齐连续切片？
   - 如何处理组织变形？

### 4.4 为什么现在是入场的最佳时机

**对研究者**：
- 技术已成熟，工具已完善
- 应用场景广泛，容易发表
- 竞争尚未饱和，仍有创新空间

**对分析师**：
- 需求爆发，就业机会多
- 技能可迁移（从 scRNA-seq 过渡容易）
- 工具生态活跃，学习资源丰富

**对临床医生**：
- 空间生物标志物有临床价值
- 病理学和分子生物学的桥梁
- 精准医疗的下一个突破口

**对投资者**：
- 市场快速增长
- 技术壁垒高，竞争格局未定
- 临床转化路径清晰

## 第五部分：证据链与边界声明

### 5.1 范式变化的证据

**证据 1：文献趋势**
- PubMed 搜索"spatial transcriptomics"：
  - 2018 年：50 篇
  - 2020 年：200 篇
  - 2023 年：1000 篇
  - 2025 年：2000+ 篇
- 增长速度超过 scRNA-seq 在 2015-2018 年的增长

**证据 2：方法学创新**
- 2020-2025 年，发表了 100+ 个 ST 分析工具
- 涵盖空间域识别、反卷积、细胞通讯、可视化等所有环节
- 工具生态的完善是技术成熟的标志

**证据 3：临床转化**
- FDA 批准的空间病理诊断工具（如 GeoMx DSP）
- 制药公司使用 ST 进行药物开发
- 空间生物标志物进入临床试验

**证据 4：教育和培训**
- 主要生信会议（如 ISMB、RECOMB）设立 ST 专题
- 在线课程和教程大量出现
- 大学开设 ST 相关课程

### 5.2 对照验证

**对照 1：scRNA-seq vs ST 的互补性**
- 用 scRNA-seq 发现细胞类型，用 ST 验证空间分布
- 用 ST 发现空间模式，用 scRNA-seq 解析细胞状态
- 两者结合是最佳策略，而非替代关系

**对照 2：ST 的独特价值**
- 负对照：打乱空间坐标后，空间模式消失
- 这证明观察到的空间模式不是偶然的
- 空间信息是不可替代的

## 总结

空间转录组不是单细胞转录组的"升级版"，而是一个新的研究范式：

**核心差异**：
- scRNA-seq：细胞是什么？
- ST：细胞在哪里、和谁对话、如何协作？

**技术成熟度**：
- 平台：从 spot 级到单细胞分辨率
- 工具：从零散到生态完善
- 应用：从探索到临床转化

**应用图谱**：
- 肿瘤学：从细胞图谱到战场地图
- 神经科学：从细胞类型到脑区地图
- 发育生物学：从时间轴到时空地图
- 免疫学：从细胞状态到免疫地形

**为什么是主战场**：
- 技术成熟，工具完善
- 应用广泛，需求旺盛
- 竞争未饱和，创新空间大
- 临床转化路径清晰

空间转录组正在改变我们理解生物学的方式。从"细胞是什么"到"细胞如何协作"，这是生物学研究的下一个前沿。
