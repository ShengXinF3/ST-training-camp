### 2.1 按技术原理分类

空间转录组平台可以分为三大类：

**测序型（Sequencing-based）**
- 原理：空间条码捕获 RNA，测序后定位
- 代表：Visium、Visium HD、Slide-seq、Stereo-seq
- 优势：全转录组检测（数万基因）
- 劣势：分辨率受限于 spot/bin 大小

**成像型（Imaging-based）**
- 原理：原位杂交或测序，直接在组织上成像
- 代表：Xenium、CosMx、MERFISH、seqFISH+
- 优势：单细胞或亚细胞分辨率
- 劣势：只能检测预设基因（几百到几千）

**ROI 型（Region of Interest）**
- 原理：选择感兴趣区域，提取 RNA 后测序
- 代表：GeoMx DSP、Curio Seeker
- 优势：灵活选择区域，全转录组
- 劣势：需要预先定义 ROI，空间连续性差

### 2.2 主流平台对比表

| Platform | Type | Resolution | Genes | Tissue Size | Sample Type | Cost |
|----------|------|------------|-------|-------------|-------------|------|
| **Visium** | Sequencing | 55μm spot | ~18,000 | 6.5×6.5 mm | FF/FFPE | $$ |
| **Visium HD** | Sequencing | 2μm bin | ~18,000 | 6.5×6.5 mm | FF | $$$ |
| **Xenium** | Imaging | Single-cell | 300-500 | 10×15 mm | FF/FFPE | $$$ |

| **Stereo-seq** | Sequencing | 500nm bin | ~18,000 | cm² scale | FF | $$ |
| **GeoMx DSP** | ROI | ROI-level | ~18,000 | 600 ROI | FF/FFPE | $$$ |
| **Slide-seq** | Sequencing | 10μm bead | ~18,000 | Customizable | FF | $ |

**缩写说明**：
- FF: Fresh Frozen（新鲜冷冻）
- FFPE: Formalin-Fixed Paraffin-Embedded（福尔马林固定石蜡包埋）
- FOV: Field of View（视野）
- ROI: Region of Interest（感兴趣区域）
- UMI: Unique Molecular Identifier（唯一分子标识符）
- NMF: Non-negative Matrix Factorization（非负矩阵分解）
- WLS: Weighted Least Squares（加权最小二乘法）
- NB: Negative Binomial（负二项分布）
- DSP: Digital Spatial Profiler（数字空间分析仪）
- BASS: Bayesian Analysis of Spatial Structure（空间结构贝叶斯分析）
- CNN: Convolutional Neural Network（卷积神经网络）
- COMMOT: COMMunication analysis by Optimal Transport（基于最优传输的通讯分析）
- LIANA: LIgand-receptor ANalysis frAmework（配体-受体分析框架）

## 第三部分：任务-工具映射表

### 3.1 核心分析任务分类

空间转录组分析可以分为 8 大任务类别：

1. **数据读取与预处理**：加载数据、格式转换
2. **质量控制**：过滤低质量 spot/细胞
3. **归一化与批次校正**：消除技术变异
4. **空间域识别**：发现功能分区
5. **空间可变基因**：识别空间模式基因
6. **细胞类型推断**：反卷积或注释
7. **细胞通讯分析**：配体-受体相互作用
8. **可视化与报告**：空间图、热图、统计图

## 第五部分：选型决策框架

### 5.1 平台选型的四个维度

**维度 1：研究问题**
- 探索性研究（不知道看什么）→ 全转录组（Visium/HD）
- 验证性研究（已知候选基因）→ 靶向平台（Xenium/CosMx）
- 细胞通讯研究 → 单细胞分辨率（HD/Xenium/CosMx）
- 组织结构研究 → 大视野（Visium/Stereo-seq）

**维度 2：样本特性**
- 新鲜冷冻 → 所有平台可选
- FFPE → Visium/Xenium/CosMx/GeoMx
- 珍贵样本（量少）→ Visium（需求量小）
- 大组织（>1cm²）→ Stereo-seq

**维度 3：资源约束**
- 预算有限 → Visium 或 Slide-seq
- 时间紧迫 → Visium（工具成熟）
- 计算资源有限 → Visium（数据量小）
- 分析能力有限 → Visium（教程多）

**维度 4：下游分析需求**
- 需要细胞类型反卷积 → 全转录组（Visium/HD）
- 需要精确细胞定位 → 单细胞分辨率（HD/Xenium/CosMx）
- 需要亚细胞结构 → 超高分辨率（Xenium/CosMx）
- 需要 3D 重建 → 连续切片（任何平台）

### 5.2 工具选型的三个原则

**原则 1：从简单到复杂**
- 先用标准方法（Scanpy + Squidpy）建立 baseline
- 遇到问题再引入专用工具
- 避免一开始就用复杂方法

**原则 2：验证工具假设**
- 每个工具都有假设（如空间平滑性、细胞类型独立性）
- 检查数据是否满足假设
- 使用诊断图验证

**原则 3：多工具对照**
- 关键结论用多个工具验证
- 比较结果的一致性
- 报告工具间的差异

## 第六部分：证据链与边界声明

### 6.1 工具选择的证据链

选择工具不是拍脑袋，需要证据支持：

**证据 1：Benchmark 研究**
- 查阅工具比较文章（如 Nature Methods 的 benchmark）
- 关注准确性、速度、稳健性指标
- 注意 benchmark 的数据类型和场景

**证据 2：方法论文**
- 阅读工具的原始论文
- 理解方法的假设和局限性
- 检查是否适用于你的数据

**证据 3：社区反馈**
- GitHub issues 和 discussions
- 论坛（如 Biostars、单细胞中文社区）
- 文献中的使用频率

**证据 4：实际测试**
- 在自己的数据上测试多个工具
- 比较结果的生物学合理性
- 评估计算成本和易用性

### 6.2 工具选择的对照验证

选择工具后，需要通过对照验证结果的可靠性：

**对照 1：多工具对照**
- 使用 2-3 个同类工具分析相同数据
- 比较核心结论的一致性
- 示例：空间域识别用 BayesSpace + SpaGCN + Leiden，看分区是否一致

**对照 2：参数敏感性**
- 改变关键参数（如聚类数、邻域半径）
- 观察结果的稳定性
- 不稳定的结果需要谨慎解释

**对照 3：负对照**
- 使用随机化数据或打乱空间坐标
- 预期不应出现空间模式
- 如果仍有"显著"结果，说明方法有问题

**对照 4：已知标志物验证**
- 用已知的组织结构或细胞类型标志物验证
- 空间域是否与组织学结构一致？
- 反卷积结果是否与 marker 基因表达一致？
